```python
clean_df.loc[43] # for printing specific rows
```

```python
# print the row for highest mid career salary
row_num = clean_df['Mid-Career Median Salary'].idxmax()
clean_df.loc[row_num]
```

```python
# print the row for lowest starting salary
row_num = clean_df['Starting Median Salary'].idxmin()
clean_df.loc[row_num]
```

```python
salary_diff = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', salary_diff) # adding a new column in dataframe
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread') # returns a new dataframe after sorting the df by the given row
low_risk.head()
```

```python
# excercise: print top 5 rows of Mid-Career 90th Percentile Salary
high_return_degrees = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending = False)
high_return_degrees.head()
```

```python
# excercise: find the degree which has highest difference in salaries
clean_df.loc[clean_df['Spread'].idxmax()]
```

```python
highest_diff = clean_df.sort_values('Spread', ascending = False)
highest_diff.head()
```
